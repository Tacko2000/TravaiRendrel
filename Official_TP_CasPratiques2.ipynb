7. Écrire un programme qui détermine pour chaque UE, le nombre d’étudiants validés. Les
résultats obtenus seront rangés dans une liste.


```python
phy = exam.loc[exam.loc[:, "Res_phy"] == "V"]
mat = exam.loc[exam.loc[:, "Res_mat"] == "V"]
chi = exam.loc[exam.loc[:, "Res_chi"] == "V"]
inf = exam.loc[exam.loc[:, "Res_inf"] == "V"]
base = exam.loc[exam.loc[:, "Res_base"] == "V"]
gr = exam.loc[exam.loc[:, "Res_gr"] == "V"]
nbre_valide = [len(phy), len(mat), len(chi), len(inf), len(base), len(gr)]
print("Le nombre d'etudiants ayant validés les differents UE respectives sont : ")
nbre_valide
```

```python
nb_chim=exam.Res_chi.value_counts()
nb_ph=exam.Res_phy.value_counts()
nb_math=exam.Res_mat.value_counts()
nb_inf=exam.Res_inf.value_counts()
nb_gr=exam.Res_gr.value_counts()
nb_ob=exam.Res_base.value_counts()
print("il y a",nb_ph[0]," étudiants ayant validé la physique")
print("il y a",nb_math[0]," étudiants ayant validé les maths")
print("il y a",nb_chim[0]," étudiants ayant validé la chimie")
print("il y a",nb_inf[0]," étudiants ayant validé l'informtique")
print("il y a",nb_ob[0]," étudiants ayant validé l'outil de base ")
print("il y a",nb_gr[0]," étudiants ayant validé l'UE graphique")

```
